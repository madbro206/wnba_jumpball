```python
# Calculate the difference in ppg between 2024 and 2023
result_df['ppg_diff'] = result_df['ppg_24'] - result_df['ppg_23']

# Filter the players who had an increase in ppg
result_df_increased = result_df[result_df['ppg_diff'] > 0]

# Sort the DataFrame by the ppg difference in descending order
result_df_increased_sorted = result_df_increased.sort_values(by='ppg_diff', ascending=False)

# Display the resulting DataFrame
print(result_df_increased_sorted[['Player', 'ppg_23', 'ppg_24', 'ppg_diff']])
```

```python
# Calculate ws/g for each player in 2023 and 2024
result_df['ws/g_23'] = result_df['ws_23'] / result_df['G_23']
result_df['ws/g_24'] = result_df['ws_24'] / result_df['G_24']

# Calculate the difference in ws/g between 2024 and 2023
result_df['ws/g_diff'] = result_df['ws/g_24'] - result_df['ws/g_23']

# Filter the players who had an increase in ws/g
result_df_increased_ws = result_df[result_df['ws/g_diff'] > 0]

# Sort the DataFrame by the ws/g difference in descending order
result_df_increased_ws_sorted = result_df_increased_ws.sort_values(by='ws/g_diff', ascending=False)

# Display the resulting DataFrame
print(result_df_increased_ws_sorted[['Player', 'G_23', 'ws/g_23', 'G_24', 'ws/g_24', 'ws/g_diff']])
```
