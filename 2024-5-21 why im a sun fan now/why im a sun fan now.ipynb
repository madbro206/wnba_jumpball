```python
import altair as alt
import pandas as pd
from IPython.display import display
import io
```

```python
df = pd.read_csv('~/Desktop/triple_doubles.csv')
```

```python
# Create the bar chart
chart = alt.Chart(df).mark_bar(color='#F88158').encode(
    x=alt.X('Team', sort='-y', title='Team'),
    y=alt.Y('count()', title='Triple doubles',            
            scale=alt.Scale(domain=[0, 12]),  # Set the y-axis domain
            axis=alt.Axis(tickMinStep=1)      # Ensure ticks at every 1 step
           )
).properties(
        title={
        "text": "Triple Doubles per WNBA Franchise (as of 5/21/2024)",
        "subtitle": ["data from acrossthetimeline.com"]
    }
)

# Display the chart in Jupyter Notebook
display(chart)
```

```python
# Define the CSV data as a multi-line string
csv_data = """Team,G,TOV,
Minnesota Lynx,2,22.0
Las Vegas Aces,2,10.5
Atlanta Dream,2,13.0
Connecticut Sun,3,11.0
New York Liberty,4,15.8
Phoenix Mercury,2,13.5
Los Angeles Sparks,2,14.0
Chicago Sky,2,14.5
Dallas Wings,2,20.0
Seattle Storm,4,18.8
Washington Mystics,3,16.0
Indiana Fever,4,17.5"""

# Use pd.read_csv() to read the CSV data as a DataFrame
df2 = pd.read_csv(io.StringIO(csv_data))

# Convert the 'TOV' column to numeric
df2['TOV'] = pd.to_numeric(df2['TOV'], errors='coerce')

# Add a new column 'TOV/G' for turnovers per game
#df2['TOV/G'] = df2['TOV'] / df2['G']

chart2 = alt.Chart(df2).mark_bar(color="#74618F").encode(
    x=alt.X('Team', sort='-y'),
    y='TOV'
).properties(
    title={
        "text": "WNBA Team TOV per Game (as of 5/20/2024)",
        "subtitle": ["data from basketball-reference.com"]
    }
)

# Display the chart in Jupyter Notebook
display(chart2)
```
