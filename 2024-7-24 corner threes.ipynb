```python
import pandas as pd
import altair as alt
```

```python
data = {
    'year': [2024, 2023, 2022, 2021, 2020, 2019],
    'corn_pct_3PA': [0.196, 0.189, 0.175, 0.153, 0.112, 0.146],
    'corn_fg_pct': [0.355, 0.378, 0.372, 0.379, 0.379, 0.361]
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
import altair as alt

# Create the scatter plot
scatter_plot = alt.Chart(df).mark_point().encode(
    x=alt.X('year:O', title='Season', axis=alt.Axis(labelAngle=-45), scale=alt.Scale(domain=[2019, 2020, 2021, 2022, 2023, 2024])),
   y=alt.Y('corn_pct_3PA', title='Percentage of 3PA that came from the corner', axis=alt.Axis(format='.1%'), scale=alt.Scale(domain=[0, 0.25]))   # Changed format to '.1%'
).properties(
    title='WNBA Percentage of 3PA from the Corner (2019-present)',
    width=400  # Adjust the width of the chart
)

# Create a line to connect the dots
line = alt.Chart(df).mark_line().encode(
    x=alt.X('year:O'),
    y=alt.Y('corn_pct_3PA')
)

# Create text labels
text = scatter_plot.mark_text(
    align='left',
    baseline='middle',
    dx=7,  # Adjust the position of the labels
    dy=3
).encode(
    text=alt.Text('corn_pct_3PA:Q', format='.1%')  # Changed format to '.1%'
)

# Combine the scatter plot, line, and text labels
scatter_plot_with_labels = scatter_plot + line + text

# Display the plot
scatter_plot_with_labels
```
