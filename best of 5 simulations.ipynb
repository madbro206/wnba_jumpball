```python
#liberty vs aces finals simulation using herhoopstats win probabilities
#generated by chatgpt

import random

def simulate_game(prob_team1_wins):
    """
    Simulate a single game between two teams.
    :param prob_team1_wins: List of probabilities of Team 1 winning each game.
    :return: True if Team 1 wins, False otherwise.
    """
    return random.random() < prob_team1_wins

def simulate_series(prob_team1_wins, num_games=5):
    """
    Simulate a best of five series between two teams.
    :param prob_team1_wins: List of probabilities of Team 1 winning each game.
    :param num_games: Number of games in the series (default is 5).
    :return: True if Team 1 wins the series, False otherwise.
    """
    team1_wins = 0
    team2_wins = 0
    
    for game in range(num_games):
        if simulate_game(prob_team1_wins[game]):
            team1_wins += 1
        else:
            team2_wins += 1
        
        if team1_wins > num_games // 2:  # Team 1 wins series when they reach more than half of the games.
            return True
        elif team2_wins > num_games // 2:  # Team 2 wins series when they reach more than half of the games.
            return False
    
    # If no team wins more than half the games, the series is tied, you can handle it accordingly.
    return None

# List of probabilities of Team 1 winning each game (adjust these values as needed).
prob_team1_wins_games = [0.75, 0.75, 0.629, 0.629, 0.75]  # Example probabilities, change as per your scenario.

# Number of simulations
num_simulations = 10000

team1_series_wins = 0
team2_series_wins = 0

for _ in range(num_simulations):
    if simulate_series(prob_team1_wins_games):
        team1_series_wins += 1
    else:
        team2_series_wins += 1

print(f"Aces series wins: {team1_series_wins}")
print(f"Liberty series wins: {team2_series_wins}")
```
