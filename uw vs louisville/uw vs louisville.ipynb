```r
#read ncaaw d1 data through 12/20/2023
data <- read.csv("team_data.csv", header=TRUE)
```

```r
#calculate points allowed
data$pts_allowed <- data$Opp/data$G
```

```r
#sort by points allowed
sorted_data <- data[order(data$pts_allowed), ]
sorted_data
```

```r
library(ggplot2)

#scatter plot pts allowed vs win percentage
ggplot(data, aes(x = pts_allowed, y = Pct)) +
  geom_point() +
  theme_bw() +
  labs(x = "Points Allowed Per Game", y = "Win Percentage", title="Points Allowed vs Win Percentage", subtitle="NCAA Women's Basketball D1 Through 12/20/2023") +
  ylim(0, 1) +
  xlim(40, max(data$pts_allowed)) +
  geom_point(data = subset(data, School == "Washington"), color = "purple", size = 3) +
  geom_text(data = subset(data, School == "Washington"), aes(label = School), color="Purple",hjust = 0, vjust = -1)
```

```r
cor(data$pts_allowed, data$Pct)
```

```r
#scatter plot pace vs win percentage
ggplot(data, aes(x = Pace, y = Pct)) +
  geom_point() +
  theme_bw() +
  labs(x = "Pace", y = "Win Percentage", title="Pace vs Win Percentage", subtitle="NCAA Women's Basketball D1 Through 12/20/2023") +
  ylim(0, 1) +
  xlim(55, max(data$pts_allowed)) +
  geom_point(data = subset(data, School == "Washington"), color = "purple", size = 3) +
  geom_text(data = subset(data, School == "Washington"), aes(label = School), color="Purple",hjust = 0, vjust = -1)
```

```r
cor(data$Pace, data$Pct)
```

```r
#scatter pace vs pts allowed
ggplot(data, aes(x = Pace, y = pts_allowed)) +
  geom_point() +
  theme_bw() +
  labs(x = "Pace", y = "Points Allowed", title="Pace vs Points Allowed", subtitle="NCAA Women's Basketball D1 Through 12/20/2023") +
  ylim(40,90) +
  xlim(55, max(data$Pace)) +
  geom_point(data = subset(data, School == "Washington"), color = "purple", size = 3) +
  geom_text(data = subset(data, School == "Washington"), aes(label = School), color="Purple",hjust = 0, vjust = -1)

cor(data$Pace,data$pts_allowed)
```
