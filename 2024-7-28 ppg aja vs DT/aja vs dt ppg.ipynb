```python
import pandas as pd
import altair as alt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load the CSV file
df = pd.read_csv('points_vs_games.csv')

# Convert game_date to datetime
df['game_date'] = pd.to_datetime(df['game_date'])

# Create a slider widget
slider = widgets.IntSlider(min=df['game_number'].min(), max=df['game_number'].max(), step=1, value=df['game_number'].min(), description='Game Number')

# Define a function to update the chart based on the slider value
def update_chart(game_number):
    filtered_df = df[df['game_number'] <= game_number]
    chart = alt.Chart(filtered_df).mark_line().encode(
        x=alt.X('game_number:Q', title='Game Number', scale=alt.Scale(domain=[0, 34])),
        y=alt.Y('cumulative_points:Q', title='Cumulative Points', scale=alt.Scale(domain=[0, 860])),
        color='athlete_display_name:N'
    ).properties(
        title='Cumulative Points by Game Number',  # Set the chart title
        width=400,
        height=400
    )

    # Clear previous output and display the updated chart
    clear_output(wait=True)
    display(slider)
    display(chart)

# Display the initial chart
update_chart(slider.value)

# Attach the update function to the slider
slider.observe(lambda change: update_chart(change['new']), names='value')

# Display the slider
display(slider)
```
