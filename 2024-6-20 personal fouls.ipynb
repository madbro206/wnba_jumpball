```python
import altair as alt
import pandas as pd

# Assuming df2 is already defined

# Line chart with points
line_chart = alt.Chart(df2).mark_line(point=alt.OverlayMarkDef(color='#2A2529')).encode(
    x=alt.X('season:O', title='Season', axis=alt.Axis(labelFontSize=12, titleFontSize=14)),
    y=alt.Y('avg_pf_per_game:Q', title='Average PF per Game', axis=alt.Axis(labelFontSize=12, titleFontSize=14)),
    tooltip=['season', 'avg_pf_per_game']
).properties(
    title='WNBA Average Fouls per Game by Season',
    width=600,
    height=400
)

# Adding text labels with better positioning
text = alt.Chart(df2).mark_text(
    align='center',
    baseline='bottom',
    dy=-10  # Adjust the vertical position of the text
).encode(
    x=alt.X('season:O'),
    y=alt.Y('avg_pf_per_game:Q'),
    text=alt.Text('avg_pf_per_game:Q', format='.1f')
)

# Combine the line chart and text
final_chart = (line_chart + text).configure(
    background='#FCEECB'  # Setting the background color
)

final_chart
```

```python
import statsmodels.api as sm

# Prepare the data
X = df2['season']
y = df2['avg_pf_per_game']

# Add a constant term for the intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
