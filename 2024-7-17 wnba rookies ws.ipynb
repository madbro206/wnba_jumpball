```python
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Player:N', sort=alt.EncodingSortField(
        field='WS',  # Sort by the WS field
        order='ascending'  # Sort in ascending order
    ), axis=alt.Axis(title='Player')),
    y=alt.Y('WS:Q', axis=alt.Axis(title='Win Shares')),
    color=alt.condition(
        alt.datum.WS >= 0,
        alt.value('#5BB2C9'),  # Positive values
        alt.value('#F88158')      # Negative values
    )
).properties(
    title='WNBA Rookie Win Shares (as of 7/16/2024)'
).configure_axis(
    labelAngle=-45
)

chart
```

```python
# Extract last names
df['LastName'] = df['Player'].apply(lambda x: x.split()[-1])

# Create the bar chart
chart2 = alt.Chart(df).mark_bar().encode(
    x=alt.X('LastName:N', sort=alt.EncodingSortField(
        field='WS',  # Sort by the WS field
        order='ascending'  # Sort in ascending order
    ), axis=alt.Axis(title='Player', labelFontSize=14)),
    y=alt.Y('WS:Q', axis=alt.Axis(title='Win Shares')),
    color=alt.condition(
        alt.datum.WS >= 0,
        alt.value('#5BB2C9'),  # Positive values
        alt.value('#F88158')     # Negative values
    )
).properties(
    title='WNBA Rookie Win Shares (as of 7/16/2024)',
    width=600,   # Set width to 600 pixels
    height=400   # Set height to 400 pixels
).configure_axis(
    labelAngle=-45
)

chart2
```
